```python
import requests
import pandas as pd

url = "https://zoopla.p.rapidapi.com/properties/list"

querystring = {"area":"Oxford, Oxfordshire","identifier":"oxford","category":"residential","order_by":"age","ordering":"descending","page_number":"1","page_size":"40"}

headers = {"X-RapidAPI-Key": "4b3e323d03msh85ecd4449385bc3p16409bjsn06beb15e8eac",
           "X-RapidAPI-Host": "zoopla.p.rapidapi.com"
}

response = requests.request("GET", url, headers=headers, params=querystring).json()
response['listing'][0]
```

```python
df = pd.DataFrame(columns=['furnished_state','per_week', 'per_month', 'category', 'num_bedrooms', 'property_type', 'tariff_type', 'supplier_name', 'gas_cost_annual', 'gas_cost_monthly', 'elect_cost_annual', 'elct_cost_monthly'])
```

```python
for details in response['listing']:
    if details['rental_prices']['shared_occupancy'] == 'N':
        shared_occupancy=details['rental_prices']['shared_occupancy']
        per_week=details['rental_prices']['per_week']
        per_month=details['rental_prices']['per_month']
        category = details['category']
        furnished_state = details['furnished_state']
        num_bedrooms = details['num_bedrooms']
        property_type = details['property_type']
        
        url = "https://zoopla.p.rapidapi.com/properties/get-running-costs"

        querystring = {"listing_id":"56354192","category":"residential","section":"to-rent"}

        headers = {"X-RapidAPI-Key": "4b3e323d03msh85ecd4449385bc3p16409bjsn06beb15e8eac",
                   "X-RapidAPI-Host": "zoopla.p.rapidapi.com"}

        responses = requests.request("GET", url, headers=headers, params=querystring).json()
        
        tariff_type=responses['running_costs']['energy']['dual_fuel']['featured_deals'][0]['tariff']['tariff_type']
        supplier_name=responses['running_costs']['energy']['dual_fuel']['featured_deals'][0]['supplier_name']
        gas_cost_annual=responses['running_costs']['energy']['dual_fuel']['featured_deals'][0]['costs']['gas']['annual']
        gas_cost_monthly=responses['running_costs']['energy']['dual_fuel']['featured_deals'][0]['costs']['gas']['monthly']
        elect_cost_annual=responses['running_costs']['energy']['dual_fuel']['featured_deals'][0]['costs']['electricity']['annual']
        elct_cost_monthly=responses['running_costs']['energy']['dual_fuel']['featured_deals'][0]['costs']['electricity']['monthly']
        
        df=df.append({'furnished_state':furnished_state, 'per_week':per_week, 'per_month':per_month, 'category':category, 'num_bedrooms':num_bedrooms, 'property_type':property_type,
                      'tariff_type':tariff_type, 'supplier_name':supplier_name, 'gas_cost_annual':gas_cost_annual, 'gas_cost_monthly':gas_cost_monthly, 'elect_cost_annua':elect_cost_annual,
                      'elct_cost_monthly':elct_cost_monthly}, ignore_index=True)




        
        
```
